```python
from textblob import TextBlob
import pandas as pd
```

```python
path = 'Modulo4/data/film_reviews_result.csv'
df = pd.read_csv(path, sep='|')
df
```

```python
# Función para traducir, analizar sentimiento y clasificar
def clasificar_sentimiento_ia(texto):
    try:
        # TextBlob trabaja mejor en inglés, traducimos si es necesario
        blob = TextBlob(texto)
        blob_en = blob.translate(to='en')
        polaridad = blob_en.sentiment.polarity
    except:
        # Si falla la traducción, intentamos el análisis directamente
        polaridad = TextBlob(texto).sentiment.polarity
    
    return 'positivo' if polaridad > 0 else 'negativo'
```

```python
df['comentario'] = df['review_text'].apply(clasificar_sentimiento_ia)
```

```python
path_save = 'Modulo4/data/comentarios_npl.csv'
df.to_csv(path_save, index=False)
```
