```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, f1_score, confusion_matrix
from sklearn.naive_bayes import MultinomialNB
from wordcloud import WordCloud  #Mostrar gráfico de palabras
```

## Cargamos los datos

```python
opcion = int(input('Ingrese 1 para los correos en español y 2 para los correos en inglés:'))

if opcion == 1:
    path = 'data/spam.csv'
    df = pd.read_csv(path, encoding='utf-8')
elif opcion == 2:
    path = 'data/spam_or_not_spam.csv'
    df = pd.read_csv(path, encoding='ISO-8859-1')
    df.rename(columns={'email':'contenido', 'label':'spam'}, inplace=True)
    #encontramos una cadena NAN  df[df.isna().any(axis=1)]
    df['contenido'] = df['contenido'].fillna('')  #La reemplazamos por cadena vacía
else:
    print('Opción no válida')
```

```python
# Agrupamos para obtener el total de datos  (0-> ham, 1->spam)
grouped = df.groupby('spam').count()
grouped
```

```python
# Gráfico de histograma de la población
df['spam'].hist()
```

```python
# Convertimos la columna spam en un arreglo
Y = df['spam'].to_numpy()
```

```python
Y
```

## Procesamiento de Datos

### Entrenamiento

```python
# Dividimos los datos en conjuntos de entrenamiento y de prueba
df_train, df_test, Y_train, Y_test = train_test_split(df['contenido'], Y, test_size=0.2)
```

```python
df_train
```

```python
df_test
```

```python
len(Y_train)
```

### Vectorizamos

```python
vectores = CountVectorizer(decode_error='ignore')
X_train = vectores.fit_transform(df_train)
X_test = vectores.transform(df_test)
```

```python
X_train
```

```python
X_test
```

### Modelo

```python
model = MultinomialNB()
model.fit(X_train, Y_train)
```

```python
# Probamos el modelo con los datos originales
train_accuracy = model.score(X_train, Y_train)
test_accuracy = model.score(X_test, Y_test)
```

```python
# Mostramos la puntuación
print(f'El accuracy de entrenamiento es de {train_accuracy}')
print(f'El accuracy de prueba es de {test_accuracy}')
```

```python
# Probamos la predicción del modelo
P_train = model.predict(X_train)
P_test = model.predict(X_test)
```

```python
# Mostramos el ajuste del modelo predicho con f1
print(f'Train F1: {f1_score(Y_train, P_train)}')
print(f'Test F1: {f1_score(Y_test, P_test)}')
```

### Matriz de Confusión

```python
conf_matrix_train = confusion_matrix(Y_train, P_train)
conf_matrix_train
```

```python
conf_matrix_test = confusion_matrix(Y_test, P_test)
conf_matrix_test
```

```python
# Gráfico de la matriz de confusión
def plot_conf_matrix(c_m, color):
    classes = ['Ham', 'Spam']
    df_cm = pd.DataFrame(c_m, index=classes, columns=classes)
    ax = sn.heatmap(df_cm, annot=True, fmt='g', cmap=color)
    ax.set_xlabel('Predecido')
    ax.set_ylabel('Objetivo')
```

```python
color = 'Blues'  #coolwarm / viridis / Blues / Greens / Reds / magma / cividis
plot_conf_matrix(conf_matrix_train, color)
```

```python
color = 'viridis'
plot_conf_matrix(conf_matrix_test, color)
```

## WordCloud

```python
def visualize(label):
    words = ''
    for msg in df[df['spam'] == label]['contenido']:
        msg = msg.lower()
        words += msg + ' '
    wordcloud = WordCloud(width=600, height=400).generate(words)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```

```python
visualize(1)  #enviamos un 1, ya que la columna spam es 1 si el correo es spam
```

### Explicación

Este generador de nube de palabras, se crea a partir de una cadena <b style="color:blue;">words</b> vacía, la cual se va llenando cada vez que al recorrer el ciclo <b style="color:green;">for</b>, se notan con mayor frecuencia ciertas palabras en los mensaje clasificados como <b style="color:red;">spam</b>. Se convierten las palabras a minúsculas y se muestran utilizando el método <b style="color:blue;">WordCloud</b> de la librería que lleva su mismo nombre.

Entre más se repite una palabra, más grande se ve en la nube de palabras.

## Identificación de Falsos Spam

```python
# Vectorizamos la columna contenido
X = vectores.transform(df['contenido'])
# Cremos la columna de predicciones
df['predicciones'] = model.predict(X)
df
```

```python
# identificación de los falsos positivos
falso_spam = df[(df['predicciones'] == 1) & (df['spam'] == 0)]['contenido']
falso_ham = df[(df['predicciones'] == 0) & (df['spam'] == 1)]['contenido']
if falso_spam.empty:
    print('No se encontraron falsos Spam')    
else:
    print('**Falsos Spam**\n')
    for msg in falso_spam:
        print(msg[:300])

if falso_ham.empty:
    print('\n\nNo se encontraron falsos Ham')    
else:
    print('\n\n**Falsos Ham**\n')
    for msg in falso_ham:
        print(msg[:10])
    
```
